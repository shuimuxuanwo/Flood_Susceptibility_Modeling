```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import warnings

# 忽略弃用警告
warnings.filterwarnings("ignore", category=DeprecationWarning)

# 美化图像参数
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] =     16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 16

# 1. 读取数据
df = pd.read_excel(r'C:\Users\la136\Desktop\数据原始.xlsx', sheet_name='二分类')

# 2. 特征列和标签列
target_col = 'label'
features = [col for col in df.columns if col != target_col]

# 3. MinMaxScaler → [0, 1]
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)
df_scaled[target_col] = df[target_col].values

# 4. 分层抽样
df_sampled = df_scaled.groupby(target_col, group_keys=False).apply(lambda x: x.sample(min(500, len(x)), random_state=42)).reset_index(drop=True)
label_map = {0: 'Safe', 1: 'Danger'}
df_sampled[target_col] = df_sampled[target_col].map(label_map)
# 5. 画平行坐标图（美化）
plt.figure(figsize=(14, 8))

# 使用 Nature 风格配色（绿蓝可改为别的）
# colors = ["#0076AA", "#EE2A25"]  # 蓝 / 橙（经典）
colors = ["#0076BB", "#D11141"]


parallel_coordinates(
    df_sampled,
    class_column=target_col,
    color=colors,
    alpha=0.4,
    linewidth=1.2
)

# 设置 legend 样式
legend = plt.legend(title='Label', fontsize=16, title_fontsize=16, loc='upper right')
for line in legend.get_lines():
    line.set_linewidth(3.0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks()
plt.tight_layout()
plt.show()
```
