```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import warnings

# 
warnings.filterwarnings("ignore", category=DeprecationWarning)

# 
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] =     16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 16

# 1. 
df = pd.read_excel(r'C:\Users\la136\Desktop\.xlsx', sheet_name='')

# 2. 
target_col = 'label'
features = [col for col in df.columns if col != target_col]

# 3. MinMaxScaler  [0, 1]
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)
df_scaled[target_col] = df[target_col].values

# 4. 
df_sampled = df_scaled.groupby(target_col, group_keys=False).apply(lambda x: x.sample(min(500, len(x)), random_state=42)).reset_index(drop=True)
label_map = {0: 'Safe', 1: 'Danger'}
df_sampled[target_col] = df_sampled[target_col].map(label_map)
# 5. 
plt.figure(figsize=(14, 8))

#  Nature 
# colors = ["#0076AA", "#EE2A25"]  #  / 
colors = ["#0076BB", "#D11141"]


parallel_coordinates(
    df_sampled,
    class_column=target_col,
    color=colors,
    alpha=0.4,
    linewidth=1.2
)

#  legend 
legend = plt.legend(title='Label', fontsize=16, title_fontsize=16, loc='upper right')
for line in legend.get_lines():
    line.set_linewidth(3.0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import warnings

# 
warnings.filterwarnings("ignore", category=DeprecationWarning)

# 
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] =     16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 16

# 1. 
df = pd.read_excel(r'C:\Users\la136\Desktop\.xlsx', sheet_name='1234')

# 2. 
target_col = 'label'
features = [col for col in df.columns if col != target_col]

# 3. MinMaxScaler  [0, 1]
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)
df_scaled[target_col] = df[target_col].values

# 4. 
df_sampled = df_scaled.groupby(target_col, group_keys=False).apply(lambda x: x.sample(min(2000, len(x)), random_state=42)).reset_index(drop=True)
label_map = {1: 'Low', 2: 'Moderate',3: 'High',4: 'Extreme'}
df_sampled[target_col] = df_sampled[target_col].map(label_map)
# 5. 
plt.figure(figsize=(14, 8))

#  Nature 
# colors = ["#0076AA", "#EE2A25"]  #  / 
colors = ["#0C1BF5", "#D11141","#D30C94", "#0ABD25"]


parallel_coordinates(
    df_sampled,
    class_column=target_col,
    color=colors,
    alpha=0.4,
    linewidth=1.2
)

#  legend 
legend = plt.legend(title='Label', fontsize=16, title_fontsize=16, loc='upper right')
for line in legend.get_lines():
    line.set_linewidth(3.0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks()
plt.tight_layout()
plt.show()
```
